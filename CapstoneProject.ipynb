# Intoduction


Coronavirus disease, also known as Covid-19, is defined as a contagious disease as a result of a recently found virus [1]. The virus was first found from a wet market in Wuhan, China, in December 2019 [2]. It then spread to the world at a very rapid rate, resulting in an ongoing pandemic. By February 2021, it was stated that over 109 million were infected by the virus, leading to approximately 2.43 million deaths [2]. About 25% of the 109 million confirmed cases were reported from the United States, causing it to be the country with the most number of confirmed cases [3] 

The virus can be spread through human-to-human transmission, where an infected person is in close-contact with others [2]. The virus can be spread via air or contaminated surface [2]. Hence, in order to prevent the virus from spreading, it is important to wear a mask whenever in public, keep social distance between people as well as regularly sanitizing everything. Furthermore, certain measures were also implemented to reduce the number of confirmed cases. These include closure of non-essential retails to prevent crowded areas, restaurants and bars are only open for take away and no dine-in to reduce human-to-human contact, to lower the possibility of people getting infected in a crowded area. 

As United State being the leading county with highest number of confirmed cases in the world, this project aims to examine the coronavirus cases (Confirmed & Death cases) across states in the United States and if it is affected by either the number of stores  or type of stores around. For instance, if a state has more stores, are the confirmed cases in that state going to be higher than states with fewer stores? In terms of type of store, it aims to study what type of stores exist in the state where there is a high number of confirmed cases.

# Data

To consider the problem, the data are listed as below 
* df_latlng, a dataset obtained online that contains the centers of population by all counties in the United States. It has the following columns : State FIPS code, County FIPS code, County Name, State Name, Population 2010, Latitude and Longitude [1].
* The Foursquare API was used to obtain the most common venues of a given county across states in the United States 
* The number of coronavirus confirmed and coronavirus death cases in the United States were obtained from 2 datasets respectively from JHU CSSE in github [2].
  + Both dataset contains the number of cases across counties in all states from 01/22/2021 onwards up to date. 
* The population data of all states were also obtained online, which has the columns as below: Rank, State, 2021 Pop., 2021 Growth, 2018 Pop. 2010 Census, Growth Since 2010, % of US, Density [3]. 
* A .json file containing all coordinates of all states of the United States were found. It was used to plot a Choropleth map of number of coronavirus confirmed and death cases of states across United States [4].
